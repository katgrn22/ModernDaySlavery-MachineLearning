```
df = pd.DataFrame(columns = ["City", "Title", "URL", "Text", "Name", "Sex", "Age", "Sexual Orientation", "Services", "Services Provided For", "Location"])
```

```
i = 0
citynum = 0
for city in url_list:
    response = requests.get(city) 

    html_soup = BeautifulSoup(response.text, 'html.parser')
    results2 = html_soup.find_all('a', class_='posttitle')
    mycity = cities[citynum]
    citynum = citynum + 1
    for result in results2:
        title = result.text[2:]
        URL = 'https://www.yesbackpage.com/' + result['href']

        sub_response = requests.get(URL) 
        html_soup_sub = BeautifulSoup(sub_response.text, 'html.parser')
        text = html_soup_sub.find('div', class_='wrap').text[1:-2]
        try:
            rows = html_soup_sub.find_all('div', class_ = 'row')[1]
            text2 = rows.find_all('tr')


            name = text2[0].find_all('td')[1].text[2:]
            sex = text2[1].find_all('td')[1].text[2:]
            age = text2[2].find_all('td')[1].text[2:]
            sexualorientation = text2[3].find_all('td')[1].text[2:]
            services = text2[4].find_all('td')[1].text[2:]
            servicesprovidedfor = text2[6].find_all('td')[1].text[2:]
            location = text2[8].find_all('td')[1].text[2:]
        except:

            name = ''
            sex = ''
            age =''
            sexualorientation = ''
            services = ''
            servicesprovidedfor = ''
            location = ''



        df.loc[i] = [mycity, title, URL, text, name, sex, age, sexualorientation, services, servicesprovidedfor, location]
        i = i +1
```
